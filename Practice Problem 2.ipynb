```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('countriesNew.csv')
data.head()
```

```python
us = data[data.country == 'United States']
us.head()
```

```python
china = data[data.country == 'China']
china.head()
```

```python
plt.plot(us.year, us.population / 10**6)
plt.plot(china.year, china.population / 10**6)
plt.legend(['US', 'China'])
plt.ylabel('population in millions')
plt.xlabel('year')
plt.title('Population growth of US and China')
plt.show()
```

```python
us_growth = us.population / us.population.iloc[0] * 100
china_growth = china.population / china.population.iloc[0] * 100
```

```python
plt.plot(us.year, us_growth)
plt.plot(china.year, china_growth)
plt.legend(['US', 'China'])
plt.ylabel('population in millions')
plt.xlabel('year')
plt.title('Population growth of US and China (first year = 100)')
plt.show()
```

```python
# Import libraries
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import files

# only load the single file
uploaded = files.upload()
# csv_housefile contains the name of the first, and only file uploaded
carscsv = next(iter(uploaded.keys()))

print('User uploaded file "{name}" with length {length} bytes'.format(
      name=carscsv, length=len(uploaded[carscsv])))
```

```python
import pandas as pd

df = pd.read_csv(carscsv)
df.head()
```

```python
df.isnull().values.any()
#check for any null values, no null values so we can continue with a more accurate representation of the data in our heatmap
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.corr())
plt.show()
```
